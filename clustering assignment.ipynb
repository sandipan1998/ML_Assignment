```python
import numpy as np
from sklearn import cluster,datasets
from scipy import misc
```

```python
# getting image of array into variable X
X=misc.face(gray=True)
```

```python
import matplotlib.pyplot as plt
```

```python
from PIL import Image
```

```python
plt.figure(figsize=(10,8))
plt.imshow(X) #function to show image which are type of array
```

```python
X
```

```python
X.shape
```

```python
X_transform=X.reshape(786432,-1) ## creating single feature    ### 768*1024=786432
```

```python
X_transform
```

```python
X_transform.shape
```

```python
from sklearn.cluster import KMeans
```

```python
clst=KMeans(n_clusters=5)
```

```python
clst.fit(X_transform)
```

```python
clst
```

```python
compressed=clst.labels_
```

```python
compressed
```

```python
compressed=compressed.reshape(768,-1)
```

```python
# visualization of compressed image

plt.figure(figsize=(10,10))
plt.imshow(compressed)
plt.axis('off')
plt.show()
```

```python
#visualization of original image

plt.figure(figsize=(10,10))
plt.imshow(X)
plt.axis('off')
plt.show()
```
